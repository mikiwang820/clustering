```python
#K means clustering
import matplotlib.pyplot as plt
import scipy.io
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial import distance
```

```python
mat = scipy.io.loadmat('/Users/wangmeiqi/Files/clustering/ex7data2.mat')

def calCost(x_train, y_kmeans, cluster_centers):
    cost = 0
    for i in range(len(x_train)):
        cost += (distance.euclidean(x_train[i], cluster_centers[y_kmeans[i]]))**2
    return cost/len(x_train)
```

```python
x_train = mat['X']
print(x_train[0])
K = np.arange(10)+1
cost = []
```

```python
for i in range(len(x_train)):
    plt.scatter(x_train[i][0], x_train[i][1], color='blue')
```

```python
for i in range(len(K)):
    kmeans = KMeans(n_clusters=K[i])
    kmeans.fit(x_train)
    y_kmeans = kmeans.predict(x_train)
    cost.append(calCost(x_train, y_kmeans, kmeans.cluster_centers_))
```

```python
plt.plot(K, cost)
min_cost = min(cost)
plt.xlabel("cost")
plt.ylabel("K")
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x_train)
y_kmeans = kmeans.predict(x_train)
y_kmeans = kmeans.predict(x_train)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_kmeans, s=50, cmap='viridis')
```
